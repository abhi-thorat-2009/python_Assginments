**1. Discuss string slicing and provide examples.**

String slicing in Python allows you to extract a part of a string by specifying a start and end index.

The format for string slicing is string[start:end], where start is the index where slicing begins (inclusive) and end is the index where it ends (exclusive).

For example, if you have the string my_string = "Hello, World!", and you use my_string[7:12], it will return "World".

```python
# example of string slicing
string="The End Game"
string[8:12] #start-a and end-b [a:b]
```

```python
b = "Hello, World!"
print(b[2:5])# Returns the characters from position 2 to position 5 (not included)
print(b[:5])# Slices from the start
print(b[2:])# Slices to the end
```


**2.Explain the key features of lists in Python.**

Lists are one of the most versatile and commonly used data structures in Python. Here are their key features -

* Ordered -

The items in a list are stored in a specific order, and that order is maintained. This means you can access elements using their index.

* Mutable -

You can modify a list after creating it. This includes adding, removing, and changing elements.

* Heterogeneous -

Lists can store elements of different data types, such as integers, floats, strings, and even other lists.

* Indexed -

Elements in a list can be accessed using their index, starting from 0 for the first element. Negative indexing is also supported, where -1 refers to the last element, -2 to the second-to-last, and so on.

* Iterable -

You can iterate over the elements in a list using loops, such as for loops.

* Dynamic -

Lists can grow or shrink in size as needed. You can add or remove elements dynamically.

* Nestable -

Lists can contain other lists as elements, allowing for the creation of complex, multi-dimensional data structures.

* Duplicates Allowed -

Lists can store duplicate values.

```python
my_list = [1, "hello", 3.14, [10, 20]]
print(my_list[1])
my_list.append("world")  # Add an element
my_list[0] = 5  # Modify an element
print(my_list)
```

**4.Compare and contrast tuples and lists the examples.**

In Python, the main difference between lists and tuples is that lists are mutable, while tuples are immutable.

* Lists -

  Can be modified after creation, such as by adding or removing elements. Lists can have variable size. For example, you can use the append() method to add an element to a list.

  List is a group of comma-separated values within square brackets and square brackets are mandatory. Eg: i = [10, 20, 30, 40]

* Tuples -

  Cannot be changed after creation. Tuples have a fixed size and do not support item assignment. For example, you cannot use the append() method on a tuple.

  Tuple is a group of comma-separated values within parenthesis and parenthesis is optional. Eg: t = (10, 20, 20, 40)


* Here are some other differences between lists and tuples -

  1) Memory usage - Tuples use less memory than lists.

  2) Speed - Tuples are generally faster than lists, especially when looking up values.

  3) When to use - Lists are used for data that needs to be changed frequently, while tuples are used for data that doesn't need to be changed frequently.



**5. Describe the key features of sets and provide examples of their use.**

##Key features of sets -

* Sets are unordered.
* Set elements are unique. Duplicate elements are not allowed.
* A set itself may be modified, but the elements contained in the set must be of an immutable type.

## uses of sets -

  Set is a data type in python used to store several items in a single variable. It is one of the four built-in data types (List, Dictionary, Tuple, and Set) having qualities and usage different from the other three. It is a collection that is written with curly brackets and is both unindexed and unordered.

* Set in python have various methods. We have seen the use of some of the above. Below is a list of all the available methods for the set objects-

  update()- used to update the set with union of others and itself

  add()- used to add a single item to the set

  copy()- used to return a copy of the set

  clear()- used to remove all items of the set

  discard()- used to remove an item from the set. If the item is not an element, then nothing is done

  union()- used to return a new set as a union of sets

  difference()- used to return a new set as the difference of two or more sets

  difference_update()- used to remove intersecting items from this set

  intersection()- used to return a new set as intersection of two sets

  intersection_update()- used to update a set with the intersection of another set and itself

  pop()- used to return and remove an arbitrary set item, KeyError is raised if the set is empty

  remove()- used to remove an item from the set. KeyError is raised if an item is not a member of the set

  issubset()- if another set is contained in this set, return true

  issuperset()- if this set is contained in another set, return true

  isdisjoint()- if the intersection of two sets is null, return true

  symmetric_difference- used to return a new set as the symmetric difference of two sets

  symmetric_difference_update()- used to update a set with the symmetric difference of another set and itself

```python
x = {"pear", "grapes", "kiwi"}

x.add("orange") # add element into set

print(x)

y = {"mango", "banana", "apple"}

x.update(y) # update value with anoter set

print(x)

x = y.copy() # copy set

print(x)

x.clear() # clear the set

print(x)

y.discard("banana") # discard the elements from set

print(y)

z = x.union(y) # union of two sets

print(z)

z = x.difference(y) # difference of two sets

print(z)
```

**6.Discuss the use cases of tuple and sets in Python programming.**

* Uses cases of tuple -

  * Storing Coordinates: Tuples are perfect for storing coordinates in 2D or 3D space, as coordinates are essentially an ordered pair or set of values.

  * Returning Multiple Values from a Function: Functions can return multiple values conveniently as a tuple.

  * Representing Records: Tuples can represent records with fixed fields, such as a student record with (name, age, grade).

* uses cases of sets -

  * Removing Duplicates:

   automatically eliminate duplicates, making them useful for cleaning data.
  * Membership Testing:

    Checking if an element is in a set is very fast, making sets ideal for tasks like filtering or searching.
  * Mathematical Operations:

    Sets can be used to perform set operations like finding common elements between two lists (intersection) or combining multiple lists without duplicates (union).

* Choosing Between Tuples and Sets:
    * Use tuples when:

      You need to store ordered data that shouldn't be changed, or when you need to use the data as a dictionary key.
    * Use sets when:

      You need to store unique elements, perform set operations, or efficiently check for membership.

**8. Discuss the importance of dictionary keys being immutable and provide example.**

In Python, dictionary keys must be immutable because they cannot be updated during runtime. Immutable objects, like strings, numbers, and tuples, have a constant hash value, which ensures that the key-value mapping is consistent. This makes it possible for Python to optimize memory usage, and it also provides inherent thread safety and predictability.

If a dictionary key were mutable, its value could change, which would also change its hash. For example, if a list were used as a key, it could contain a reference to itself, which would cause the copying code to run into an infinite loop. Allowing lists as keys could also lead to hard-to-track bugs if a list were accidentally modified

Here are some examples of immutable types that can be used as dictionary keys in Python:

String, Number, Tuple, Integer, Float, and Boolean.

```python
# example of dict
my_dict={"name":"Ram","age":45} # here keys are immutable
my_dict
```
