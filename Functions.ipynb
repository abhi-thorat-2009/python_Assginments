1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.



```python
l=input("Enter a list of numbers with comma sepater (,) - ")
l=l.split(",")
sum=0
for i in l:
  if int(i)%2==0:
    sum+=int(i)
print("Sum of all even numbers in list is - ",sum)
```

3, Implement a Python function that takes a list of integers and returns a new list containing the squares of each number,



```python
l=input("Enter a list of numbers with comma sepater (,) - ")
l=l.split(",")

def square(l):
  for i in range(len(l)):
    l[i]=int(l[i])**2
  return l

print("Squares of each number in list is - ",square(l))

```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.



```python
def prime(n):
    for i in range(2,n):
      if n%i==0:
        return False
    return True


n=int(input("Enter a number  from 1 to 200 - "))
if n<=200 and n>=1:
  if prime(n):
    print(n,"is a prime number")
  else:
    print(n,"is not a prime number")
else:
  print("Number should be from 1 to 200")


```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.



```python
def fibonacci(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b

n=int(input("Enter a number where you want fibonacci series - "))

__iter__=fibonacci(n) # creating iterator of class fibonacci
for i in range(n):
  print(next(__iter__),",",end="  ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python

def power(n):
  for i in range(n):
    yield 2**i


n=int(input("Enter a number where you want powers of 2 - "))

__iter__=power(n) # creating iterator of class power
for i in range(n):
  print(next(__iter__),",",end="  ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
# Implement a generator function that reads a file line by line and yields each line as a string.

def read_file(file_name):
  with open(file_name,"r") as f:
    for line in f:
      yield line

```

9. Write a Python program that uses 'map() to convert a list of temperatures from Celsius to Fahrenheit.



```python
l=input("Enter a list of temperatures in celsius  with comma sepater (,) - ")
l=l.split(",")
l=list(map(lambda x:int(x)*9/5+32,l)) #(0°C × 9/5) + 32 = 32°F this formla to convrt
print("list of temperature in fahrenheit - ",l)
```

10. Create a Python program that uses 'filter() to remove all the vowels from a given string.

```python
string=input("Enter a string - ")
vowels=["a","e","i","o","u","A","E","I","O","U"]
string1=list(filter(lambda x:x not in vowels,string))
print("String without vowels - ","".join(string1))
```
