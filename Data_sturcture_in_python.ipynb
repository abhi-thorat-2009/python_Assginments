4. Write a code to check if two given strings are anagrams of each other.

```python
# 1st step is check for length should be same and convert both strings into lower or upper case
# 2nd step is take alpebatical order of each string and compare it if it same then it is a anagrams of each other
s1="Race"
s2="care"
if len(s1)==len(s2):
  s1=s1.lower()
  s2=s2.lower()
  if sorted(s1)==sorted(s2):
    print("both strings are anagrams of each other")
  else:
    print("both strings are not anagrams of each other")
else:
  print("both strings are not anagrams of each other ")
```

6. Write a code to perform basic string compression using the counts of repeated characters.



```python
def string_compression(s):
    result = ""
    if not s:
        return result
    char_count = 1  # Initialize character count to 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            char_count += 1
        else:
            result += s[i - 1] + str(char_count)
            char_count = 1
    result += s[-1] + str(char_count)
    return result

string="aaasserterrrdddggf"
print("the string compression of counts of repeated characters for this string - ",string)
print("string compression is - ",string_compression(string))
```

32. Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together.



```python
dict1 = {'a': 10, 'b': 35, 'c': 30}
dict2 = {'b': 15, 'd': 40,'c':40}
merged_dict = dict1.copy()
for key, value in dict2.items():
    if key in merged_dict:
      merged_dict[key] += value
    else:
      merged_dict[key] = value
print(merged_dict)
```

34. Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.



```python
my_dict = {'apple': 5, 'banana': 2, 'cherry': 8}

# Sort in ascending order based on values
sorted_dict_asc = dict(sorted(my_dict.items(), key=lambda item: item[1]))
print("Ascending order:", sorted_dict_asc)

# Sort in descending order based on values
sorted_dict_desc = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))
print("Descending order:", sorted_dict_desc)
```

35. Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary

```python
my_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3}

inverted_dict = {}
for key, value in my_dict.items():
  if value not in inverted_dict:
    inverted_dict[value] = [key]
  else:
    inverted_dict[value].append(key)

print(inverted_dict)
```
